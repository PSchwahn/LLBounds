```python
V = FiniteRankFreeModule(QQ, 7, name='V', start_index=0)
e = V.basis('e')
e
```

```python
ed=e.dual_basis()
ed
```

```python
phi=ed[0].wedge(ed[1].wedge(ed[2])+ed[3].wedge(ed[4])+ed[5].wedge(ed[6]))+ed[1].wedge(ed[3].wedge(ed[5])-ed[4].wedge(ed[6]))-ed[2].wedge(ed[3].wedge(ed[6])+ed[4].wedge(ed[5]))
phi
```

```python
gdown = V.tensor((2,0), name='g')
for i in range(0,7):
    gdown[i,i]=1
gdown.disp()
```

```python
phi2=phi['_i..']*gdown['^i.']
u=phi['_i..']*phi2['^i_..']
u
```

```python
R=PolynomialRing(QQ,49,"x")
x=R.gens()
```

```python
P=[0 for n in range(0,49)]
```

```python
for i in range(0,7):
    for j in range(0,7):
        ij=i%7*7+j
        for p in range(0,7):
            for q in range(0,7):
                pq=p%7*7+q
                for r in range(0,7):
                    for s in range(0,7):
                        rs=r%7*7+s
                        P[ij]+=phi[i,p,r]*phi[j,q,s]*x[pq]*x[rs]
```

```python
P[48]
```

```python
I=ideal(P[n] for n in range(0,49))#bis 8 geht es. bis 49 kommen wir wohl nicht...
B=I.groebner_basis()
B[0]
```

```python
Q=0
for i in range(0,7):
    for j in range(0,7):
        ij=i%7*7+j
        for p in range(0,7):
            for q in range(0,7):
                pq=p%7*7+q
                for r in range(0,7):
                    for s in range(0,7):
                        rs=r%7*7+s
                        Q+=phi[i,p,r]*phi[j,q,s]*x[pq]*x[rs]*x[ij]
```

```python
Q
```

```python
Proj = ProjectiveSpace(48, QQ, 'x')
k=49#k=8 works. can we crank this up to 49?
Partial=[P[n] for n in range (0,k)]
Var=Proj.subscheme(Partial)
48-k-Var.dimension()
```
